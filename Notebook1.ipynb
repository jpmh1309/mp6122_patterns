## What is PCA?

Principal Components Analysis (PCA) is arguably one of the most widely used statistical methods. It has applications in nearly all areas of statistics and machine learning including clustering, dimensionality reduction, face recognition, signal processing, image compression, visualisation and prediction.

PCA is based on the computation of eigenvalues and eigenvectors. Basically, PCA defines a new orthogonal coordinate system that optimally describes variance in a single dataset. PCA is based on the concept of finding the vectors or dimensions that provide more information in the distribution of the observed data. This means finding the **axes** that imply a greater variation of the data with respect to each one of them. The **axes** with low variance with respect to the data imply a small contribution of information, so they can be discarded.

To find such **axes**, PCA proposes the decomposition into eigenvalues and eigenvectors of the covariance matrix $\sum$ associated with the data matrix $A$. The covariance matrix $\sum$ for data of dimensionality $n$ is of dimensions $n\times n$ with the diagonal specifying the variation of the data in each dimension, and in inputs other than the diagonal defining the variation between the pair of specified dimensions.

The **axes** exemplified by corresponding to the eigenvectors of the covariance matrix have associated eigenvalues, which define the contribution of their eigenvector to the data distribution. In this way it is possible to reduce the dimensionality of the data, creating a new vector space which includes the eigenvectors with the largest values of their eigenvalues

So, the two mainly uses of PCA are: 

* Feature Elimination: PCA takes data in high dimensions and create a reduced-representation of the data.

* Feature Extraction: PCA helps to extract meaningful factors from high dimensional data in a way that helps us interpret the major trends in our data.

### When should I use PCA?

1. Do you want to reduce the number of variables, but aren’t able to identify variables to completely remove from consideration?
2. Do you want to ensure your variables are independent of one another?
3. Are you comfortable making your independent variables less interpretable?

If you answered “yes” to all three questions, then PCA is a good method to use. If you answered “no” to question 3, you should not use PCA.

## PCA in "eigenfaces"

The Principal Component Analysis (PCA) is one of the most successful techniques that have been used in image recognition and compression. In 1991, Turk and Pentland suggested an approach to face recognition that uses dimensionality reduction and linear algebra concepts to recognize faces. This approach is computationally less expensive and easy to implement and thus used in various applications at that time such as handwritten recognition, lip-reading, medical image analysis, etc. 

Basically, eingenfaces is the name given to a set of eigenvectors when used in the computer vision problem of human face recognition. 

A set of eigenfaces can be generated by performing a PCA on a large set of images depicting different human faces. The eigenfaces themselves form a basis set of all images used to construct the covariance matrix. This produces dimension reduction by allowing the smaller set of basis images to represent the original training images. Classification can be achieved by comparing how faces are represented by the basis set.

Informally, eigenfaces can be considered a set of "standardized face ingredients", derived from statistical analysis of many pictures of faces.

Most commonly used for feature extraction in pattern classification problem

**References**: 

- https://www.apsl.net/blog/2017/07/18/using-linear-discriminant-analysis-lda-data-explore-step-step/

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

%matplotlib inline

wine = datasets.load_wine()
X = wine.data
y = wine.target
target_names = wine.target_names

X_r_lda = LinearDiscriminantAnalysis(n_components=2).fit(X, y).transform(X)
X_r_pca = PCA(n_components=2).fit(X).transform(X)

with plt.style.context('seaborn-talk'):
    fig, axes = plt.subplots(1,2,figsize=[15,6])
    colors = ['navy', 'turquoise', 'darkorange']
    for color, i, target_name in zip(colors, [0, 1, 2], target_names):
        axes[0].scatter(X_r_lda[y == i, 0], X_r_lda[y == i, 1], alpha=.8, label=target_name, color=color)
        axes[1].scatter(X_r_pca[y == i, 0], X_r_pca[y == i, 1], alpha=.8, label=target_name, color=color)
    axes[0].title.set_text('LDA for Wine dataset')
    axes[1].title.set_text('PCA for Wine dataset')
    axes[0].set_xlabel('Discriminant Coordinate 1')
    axes[0].set_ylabel('Discriminant Coordinate 2')
    axes[1].set_xlabel('PC 1')
    axes[1].set_ylabel('PC 2')
```

## Data Imputation

The imputation methods consist of estimating the missing values based on the valid values of other variables and/or cases in the sample. The estimation can be made from the information of the complete set of variables or from some specially selected variables. Imputation methods are usually used with metric variables (interval or ratio), and must be applied with great caution because they can introduce non-existent relationships in the real data.

As an example, we have that: 

| Sample | Data1 | Data2 | Data3 | Data4 | Data5 |

| **0**  | 2.0   | 5.0   | 3.0   | 6.0   | NaN   |
| **1**  | 9.0   | NaN   | 9.0   | 0.0   | 7.0   |
| **2**  | 19.0  | 17.0  | NaN   | 9.0   | NaN   |
| **3**  | 7.0   | 10.0  | 3.0   | 6.0   | 4.0   |
| **4**  | 2.0   | 8.0   | 10.0. | NaN   | 3.0   |

Is converted to:

| Sample | Data1 | Data2 | Data3 | Data4 | Data5 |
|--------|-------|-------|-------|-------|-------|
| **0**  | 2.0   | 5.0   | 3.0   | 6.0   | 4.67  |
| **1**  | 9.0   | 10.0  | 9.0   | 0.0   | 7.0   |

Some of the most common methods are: 

### k-Nearest Neighbor Imputation
This method uses the closest value criterion associated with an auxiliary variable to provide the imputed or donor values. In the event that the auxiliary variable has several equidistant values, the problem of the presence of several donors for the same missing value arises. In this situation, the criterion that is followed is to randomly choose a donor among the different candidates

### Random Hot Deck
This method consists of selecting by simple random sampling with replacement $m$ values from the $r$ available values of the variable $y$.

### Mean Substitution
It consists of substituting the missing value for the mean of the valid values.

### Constant Substitution
It consists of replacing the missing values with constants whose value is determined for theoretical reasons or reasons related to previous research. It has the same drawbacks as mean substitution, and should only be used if there are reasons to believe that it is more suitable than the mean method.

### Regression Imputation
This method consists of estimating the missing values based on their relationship with other variables through regression analysis.

## Data Balancing

A dataset is imbalanced if the classification categories are not approximately equally represented.

balance.png

#### Why data balancing is important?

Let’s consider the above example, where we had class A with 90 observations and class B with 10 observations. If we predict the entire data set as class A, we will achieve an accuracy of 90% which seems really not bad for a classification model. But, in reality this model is very poor. To expand on this further, if we consider a quality control process and the model predicts everything as good, then we end up getting a lot of warranty claims.

There are some tecniques that can be used to solve this problem. For example:

### Oversampling

This technique is used to modify the unequal data classes to create balanced datasets. When the quantity of data is insufficient, the oversampling method tries to balance by incrementing the size of rare samples.

Upsampling or oversampling, increases the number of minority class members in the training set. The advantage of over-sampling is that no information from the original training set is lost, as all observations from the minority and majority classes are kept. On the other hand, it is prone to over fitting

### Downsampling

This technique balances the imbalance dataset by reducing the size of the class which is in abundance. 

Downsampling or Undersampling, on contrary to oversampling, aims to reduce the number of majority samples to balance the class distribution. Since it is removing observations from the original data set, it might discard useful information.

### SMOTE

Synthetic Minority Over-sampling Technique (SMOTE) is a  method that makes up some data — but in a good way. That “synthetic” part refers to a process by which additional data similar to those in your minority class are generated and can be added to your dataset to bring the classes into balance.

SMOTE works by choosing one of your observations representing the minority class, finding its nearest neighbors (you can specify how many), and using the relationship between the chosen observation and the neighbor(s) to generate a new example of the minority class.

### ADASYN

Adaptive Synthetic Sampling Method for Imbalanced Data (ADASYN) is a sampling approach for learning from imbalanced data sets. The essential idea of ADASYN is to use a weighted distribution for different minority class examples according to their level of difficulty in learning, where more synthetic data is generated for minority class examples that are harder to learn compared to those minority examples that are easier to learn. As a result, the ADASYN approach improves learning with respect to the data distributions in two ways: (1) reducing the bias introduced by the class imbalance, and (2) adaptively shifting the classification decision boundary toward the difficult examples. 

### SMOTE vs ADASYN

The major difference between SMOTE and ADASYN is the difference in the generation of synthetic sample points for minority data points. In ADASYN, we consider a density distribution $r_x$ which thereby decides the number of synthetic samples to be generated for a particular point, whereas in SMOTE, there is a uniform weight for all minority points.

## One Hot Encoding 

Basically, one hot encoding is a process by which categorical variables are converted into a form that could be provided to ML algorithms to do a better job in prediction. A categorical data is defined as variables with a finite set of label values.

Some machine learning algorithms can work directly with categorical data depending on implementation, such as a decision tree, but most require any inputs or outputs variables to be a number, or numeric in value. This means that any categorical data must be mapped to integers. 

One hot encoding is one method of converting data to prepare it for an algorithm and get a better prediction. With one-hot, we convert each categorical value into a new categorical column and assign a binary value of 1 or 0 to those columns. 

As an example, we have that: 

| **RED** | **GREEN** | **BLUE** |
|---------|-----------|----------|

Is converted to:

| **RED** | **GREEN** | **BLUE** |
|---------|-----------|----------|
| 1       | 0         | 1        |
| 0       | 1         | 0        |
| 0       | 0         | 1        |
